```python
class Camera: 
    def __init__(self, name):
        self.name = name
    def capture(self):
        print('A photo is captured')
```

```python
class Smart(Camera):
 def __init__(self, name, mp):
  super().__init__(name)
  self.mp = mp


class DSLR(Camera):
 def __init__(self, name, mp):
  super().__init__(name)
  self.mp = mp

class Drone(Camera):
 def __init__(self, name, mp):
  super().__init__(name)
  self.mp = mp

phone = Smart('iphone', '12MP')
dslr = DSLR('canon', '24MP')
drone = Drone('dji', '4KMP')


phone.capture()
dslr.capture()
drone.capture()
```

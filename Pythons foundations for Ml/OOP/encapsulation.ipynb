```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei

    def get_imei(self):
            return self.__imei

iphone = Mobile("Apple", "iPhone 12", "123456789012345")
print(f"Mobile Name: {iphone.name}")
print(f"Mobile Model: {iphone.model}")
print(f"Mobile IMEI: {iphone.get_imei()}")
```
